```python
# Libraries
#import sacremoses
import torch
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from simpletransformers.ner import NERModel,NERArgs
import pandas as pd
```

```python
import re
```

```python
def parse_corpus(file_path):
    """Parse a single corpus file and return a list of word-tag tuples."""
    corpus = []
    with open(file_path) as file:
        for line in file:
            # Remove trailing newline and unnecessary spaces
            line = line.strip().replace(" ", "")
            words = re.split(r"<.{2,4}>", line)
            tags = re.findall(r"<.{2,4}>", line)
            # Pair words with their tags and clean up spaces
            corpus.append([(word.strip(), tag[1:-1]) for word, tag in zip(words, tags)])
    return corpus
```

```python
def load_corpus(last=False, test=2, all=False):
    """Load and parse corpus files based on the given flags."""
    corpus = []
    file_paths = [
        "Dataset/Corpus/00ne_pos.txt",
        "Dataset/Corpus/01ne_pos.txt",
        "Dataset/Corpus/02ne_pos.txt"
    ]

    # Determine which files to load based on flags
    if all:
        selected_files = file_paths
    else:
        selected_files = [
            file_paths[0] if (last and test == 0) or (not last and test != 0) else None,
            file_paths[1] if (last and test == 2) or (not last and test != 2) else None,
            file_paths[2] if (last and test == 1) or (not last and test != 1) else None,
        ]
        # Filter out None values
        selected_files = [path for path in selected_files if path]

    # Parse the selected files
    for file_path in selected_files:
        corpus.extend(parse_corpus(file_path))

    return corpus
```

```python
tagged_sentences = load_corpus(all=True)
tagged_sentences[:2]
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Flatten the tagged_sentences into a list of dictionaries
data = []

# Convert to list of dictionaries
for sent_idx, sentence in enumerate(tagged_sentences, start=1):
    for word, pos in sentence:
        data.append({
            'sentence#': f'sentence: {sent_idx}',
            'word': word,
            'pos': pos,
        })

# Convert to a DataFrame
df = pd.DataFrame(data)
df
```

```python
# Apply LabelEncoder to "sentence#"
label_encoder = LabelEncoder()
df["sentence#"] = label_encoder.fit_transform(df["sentence#"])
df.head(5)
```

```python
#Note that 'pos' should be used instead of 'ner' for POS tagging task
df.rename(columns={"sentence#":"sentence_id","word":"words","pos":"labels"}, inplace =True)
```

```python
df.tail(5)
```

```python
df["labels"] = df["labels"].str.upper()
```

```python
X= df[["sentence_id","words"]]
Y =df["labels"]
```

```python
# Train-test split
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size =0.2, shuffle=True)
```

```python
#building up train data and test data
train_data = pd.DataFrame({"sentence_id":x_train["sentence_id"],"words":x_train["words"],"labels":y_train})
test_data = pd.DataFrame({"sentence_id":x_test["sentence_id"],"words":x_test["words"],"labels":y_test})
print(len(train_data),len(test_data))
```

#### Model Configuration

1. **NERArgs Configuration**:
   - `num_train_epochs = 5`: The model will train for 5 epochs.
   - `learning_rate = 1e-4`: Specifies the learning rate for the optimizer.
   - `overwrite_output_dir = True`: Ensures the output directory is overwritten during training.
   - `train_batch_size = 8`: Sets the training batch size to 8.
   - `eval_batch_size = 8`: Sets the evaluation batch size to 8.
   - `save_model_every_epoch = False`: Prevents saving the model after each epoch to save storage.

2. **NERModel Initialization**:
   - `model_type = 'bert'`: Specifies the base model architecture (BERT).
   - `model_name = 'NepBERTa/NepBERTa'`: Indicates the pre-trained NepBERTa model from Hugging Face.
   - `labels`: Defines the list of labels used for the NER task.
   - `args`: Passes the configuration object (`NERArgs`) to the model.
   - `use_cuda = False`: Disables GPU usage (set to `True` if GPU is available).
   - `from_tf = True`: Indicates the model is loaded from TensorFlow format (remove for other model formats).

```python
# Model arguments and Parameters
args = NERArgs()
args.num_train_epochs = 5
args.learning_rate = 1e-4
args.overwrite_output_dir =True
args.train_batch_size = 8
args.eval_batch_size = 8
args.save_model_every_epoch=False
```

```python
# Change according to model eg. 'xlmroberta', 'xlm-roberta-base'. Refer to simpletransformer docs and hugging face for more models
model = NERModel('bert', 'NepBERTa/NepBERTa',labels=label,args =args,use_cuda=False,from_tf=True)# remove 'from_tf=True' for other models
```

```python
# Training the model
model.train_model(train_data,eval_data=test_data, acc=accuracy_score, verbose=True)
```

```python
# Evaluating scores
result, model_outputs, preds_list = model.eval_model(test_data, verbose=True)
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix
import numpy as np

# Flatten the lists
actual_flat = [item for sublist in actual_labels_list for item in sublist]
predicted_flat = [item for sublist in preds_list for item in sublist]

# Get unique labels
labels = np.unique(actual_flat)

# Compute confusion matrix
conf_matrix = confusion_matrix(actual_flat, predicted_flat, labels=labels)

print("Confusion Matrix:")
print(conf_matrix)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define function to plot confusion matrix
def plot_confusion_matrix(cm, labels):
    plt.figure(figsize=(40, 40))
    sns.set(font_scale=1.2)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title('Confusion Matrix')
    plt.show()

# Plot confusion matrix
plot_confusion_matrix(conf_matrix, labels)
```
